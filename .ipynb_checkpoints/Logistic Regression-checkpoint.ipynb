# Logistic Regression Study Notes

Yes it's name includes 'Regression'; however Logistic Regression predicts the probabilities of classess therefore it is used for Classification problems.

How it works?

- Just as Linear Regression, Logistic Regression gets the weighted sum of the input features and adds a bias term. However instead of resulting exact numbers as linear regression; Logistic Regression uses Sigmoid Function to get the probability of the classess.

$$ Logistic Regression = h_\theta(x) = \sigma(x^T\theta) $$

Where $ \sigma $ is equal to:

$$ \sigma(t)  = \frac{1}{1 + e^{(-t)}} $$

The steps afterwards are easy;

- Once the LogReg predicts the probability  the following equation used to make class predictions.


$ \hat{y} $ = 0 if $\hat{p}$ < 0.05 

or

$\hat{y}$ = 1 if $\hat{p}$ >0.05 

```python
import numpy as np
from sklearn.datasets import make_classification
```

```python

class LogReg():
    def __init__(self, lr = 0.01, n_iters = 1000):
        self.lr = lr
        self.n_iters = n_iters
        self.bias = None
        self.weights = None


    def sigmoid( z):
        # convert input to a numpy array
        z = np.array(z) 
        g = np.zeros(z.shape)

        from math import e
        
        g = (1 / (1 + pow(e,-z)))
        
        return g 

    
    def cost_function(theta, X, y):
        sigmoid = LogReg.sigmoid(z = y)
        m = y.size  # number of training examples

        # You need to return the following variables correctly 
        J = 0
        grad = np.full((len(X),1),theta)

        # ====================== YOUR CODE HERE ======================
        
        J = - (1/m)* (sum((y*np.log(sigmoid(np.dot(X,theta))))+((1-y)*np.log(1-sigmoid(np.dot(X,theta))))))
        
        grad = (1/m) * ((sigmoid(X.dot(theta.T))-y).dot(X))
        
        # =============================================================
        return J, grad
```

```python
# create a binary dataset
X, y = make_classification(n_samples=1000, random_state=np.random.RandomState(0))
```

```python
LogReg.cost_function(theta=0.01, X= X,y = y)
```
